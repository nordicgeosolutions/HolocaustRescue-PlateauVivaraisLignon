```python
## Load the ENGLISH-language dataset for analysis.

path = "YOUR_DATA_Eng"

def read_txt_files(directory):
    # Reads all .txt files in a directory and returns a combined string of their contents.

    file_contents = ''
    
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            filepath = os.path.join(directory, filename)
            with open(filepath, "r", encoding="utf8") as f:
                file_contents = file_contents + (f.read())
    return file_contents

texts = read_txt_files(path)
```

```python
# Import ENGLISH emotion classification model, and "warm it up" on the simple sentence.
from transformers import pipeline
classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base")
classifier("I love this!")
```

```python
## Define function to perform emotion classification on text that is 150 characters on either side of keywords

def detect_emotion_with_context(text, keywords, context_size=150):
    """
    Detects emotions associated with keywords in text, providing surrounding context.

    Args:
        text (str): The input text.
        keywords (list): A list of keywords to search for.
        context_size (int): The number of characters to include as context 
                          around the keyword.  In this case, I used 150 (tried 50, 75, 100, 125, 175, 200, 225, 
                          250, 275, 350, but results were more accurate based on my sampling of results at 150).

    Returns:
        dict: A dictionary where keys are keywords found, and values are lists 
              of tuples. Each tuple contains:
              - The classified emotion (from hartmann_distil_roberta).
              - The surrounding context of the keyword.
    """
    results = {}
    for keyword in keywords:
        keyword_results = []
        # Case-insensitive search for the keyword with re.IGNORECASE
        for match in re.finditer(r'\b' + re.escape(keyword) + r'\b', text, re.IGNORECASE):
            start = max(0, match.start() - 150)
            end = min(len(text), match.end() + 150)
            context = text[start:end]
            emotion = classifier(context)
            
            keyword_results.append((keyword, emotion, context))
        if keyword_results:
            results[keyword] = keyword_results
    return results
```

```python

## IMPORTANT!!  Need to do each set of keywords one at a time, all the way through exporting the results
## to their respective csv file.  This is because the model is run once per set of keywords.
```

```python
# PLATEAU LOCATIONS: Run the model on the ENGLISH dataset, with Plateau Locations as the keywords in this instance, 
# and print results.

text = texts
keywords = PlatLoc

emotion_data = detect_emotion_with_context(text, keywords)

for keyword, data_list in emotion_data.items():
    #print(f"Keyword: {keyword}")
    for keyword, emotion, context in data_list:
        print(f"  Keyword: {keyword}")
        print(f"  Emotion: {emotion}")
        print(f"  Context: {context}")
```

```python
# PLATEAU LOCATIONS: Results of emotion classifications get appended to a list, then converted to a dataframe in lines 1-5. 
# Then, lines 6-8 normalize and remove the list that contained the dictionary of the classified emotion label 
# and score.  Then, lines 9-10 normalize the dictionary and create separate columns for the label and score.  
# Then, lines 11-21 adds a column for the numerical value of the emotion using the key values from the 
# Hartmann Roberta model.  Lines 22 just displays the final normalized dataframe.

results = []
for keyword, data_list in emotion_data.items():
        for keyword, emotion, context in data_list:
            results.append([keyword, emotion, context])
df_results = pd.DataFrame(results, columns = ["Plateau_Locations", "Emotion", "Context"])
df_norm = pd.json_normalize(df_results["Emotion"])
df_results_norm = pd.concat([df_results, df_norm], axis=1).drop("Emotion", axis=1)
df_results_norm.columns = ["Plateau_Locations", "Context", "Emotion"]
df_norm2 = pd.json_normalize(df_results_norm["Emotion"])
df_results_norm_final = pd.concat([df_results_norm, df_norm2], axis=1).drop("Emotion", axis=1)
conditions = [
    df_results_norm_final["label"] == "anger",
    df_results_norm_final["label"] == "disgust",
    df_results_norm_final["label"] == "fear",
    df_results_norm_final["label"] == "joy",
    df_results_norm_final["label"] == "neutral",
    df_results_norm_final["label"] == "sadness",
    df_results_norm_final["label"] == "surprise"
]
values = ["0", "1", "2", "3", "4", "5", "6"]
df_results_norm_final["label_numeric"] = np.select(conditions, values, default="Unknown")
df_results_norm_final
```

```python
# PLATEAU LOCATIONS:  This merges two dataframes: the first is the dataframe with the extracted Plateau locations 
# and their coordinates and the second is the dataframe which contains the contextualized emotions surrounding 
# references to these Plateau locations. Then it exports the results to a csv file.

df_mergedPlatLoc = pd.merge(df_PlatLoc, df_results_norm_final, on="Plateau_Locations", how="right")

df_mergedPlatLoc.to_csv("Results_EmotionsPlateauLocations.csv", encoding="utf-8-sig", index=False, header=True, mode="w+")
```

```python
# MODE of PLATEAU LOCATIONS:  Calculate the MODE of which emotion is most prevalent at a given set of coordinates...
# The first line is a repeat of the first line in the cell above, thus avoiding having to re-export the 
# primary results. Thus, it merges two dataframes: the first is the dataframe with the extracted locations 
# and their coordinates and the second is the dataframe which contains the contextualized emotions surrounding 
# references to these locations. 
# Then the next line drops any rows that have a score less than 0.50.
# Then the next line groups by the coordinates of the locations, and consolidates by the MODE of emotion (numeric),
# then prints it.  If I don't want MODE results as a list, see file CalculateModePandasDataframe.txt.
# Then the next line exports results to a csv.

df_mergedPlatLoc2 = pd.merge(df_PlatLoc, df_results_norm_final, on="Plateau_Locations", how="right")
df_mergedPlatLoc2 = df_mergedPlatLoc2.drop(df_mergedPlatLoc2[df_mergedPlatLoc2['score'] < 0.50].index)
df_mergedPlatLoc2_mode = df_mergedPlatLoc2.groupby(["LAT", "LON", "ALT"], as_index=False)["label_numeric"].agg(lambda x:list(pd.Series.mode(x)))
print(df_mergedPlatLoc2_mode)
df_mergedPlatLoc2_mode.to_csv("Results_EmotionsPlateauLocations_mode.csv", encoding="utf-8-sig", index=False, header=True, mode="w+")
```

```python
# AGGREGATION of PLATEAU LOCATIONS:  Aggregates the emotion scores in the dataframe based on the coordinates 
# and creates a list of all the scores found at that location, then normalizes/explodes the list.
# The first line is a repeat of the first line in the cell above, thus avoiding having to re-export the 
# primary results. Thus, it merges two dataframes: the first is the dataframe with the extracted locations 
# and their coordinates and the second is the dataframe which contains the contextualized emotions surrounding 
# references to these locations. 
# Then the next line aggregates the data as described above, resulting in a list in the column of all the numeric
# scores for emotion at that location.
# Then the next line explodes the numeric label for emotion into separate rows for each emotion numeral.
# Then the next line exports results to a csv.

df_mergedPlatLoc3 = pd.merge(df_PlatLoc, df_results_norm_final, on="Plateau_Locations", how="right")
df_mergedPlatLoc3_agg = df_mergedPlatLoc3.groupby(["LAT", "LON", "ALT"], as_index=False).agg({'label_numeric': lambda x: list(set(x))})
df_mergedPlatLoc3_agg = df_mergedPlatLoc3_agg.explode("label_numeric")
print(df_mergedPlatLoc3_agg)
df_mergedPlatLoc3_agg.to_csv("Results_EmotionsPlateauLocations_AGG.csv", encoding="utf-8-sig", index=False, header=True, mode="w+")
```

```python
# COUNT/PROPORTION of EMOTION SCORES of PLATEAU LOCATIONS:  Calculate the COUNT of each emotion detected at a 
# given set of coordinates then calculate the proportion at that location of each emotion...
# The first line is a repeat of the first line in the cell above, thus avoiding having to re-export the 
# primary results. Thus, it merges two dataframes: the first is the dataframe with the extracted locations 
# and their coordinates and the second is the dataframe which contains the contextualized emotions surrounding 
# references to these locations. 
# Then the next line groups by the coordinates of the locations, and COUNTS the occurrences of emotion 
# scores (in the column "label_numeric"), then the "normalize=True" argument calculates and returns the 
# relative frequency of that score.  
# The next line renames the column for "label_numeric" to "emotion_value."
# The next sequence of lines sets conditions to add the "labels" to the "emotion_value" scores.
# The next 3 lines calculates the confidence scores for the detected emotions at each location, merges the
# results with the normalized count dataframe and prints the results.
# Then the next line exports results to a csv.

df_mergedPlatLoc4 = pd.merge(df_PlatLoc, df_results_norm_final, on="Plateau_Locations", how="right")
df_mergedPlatLoc4_count = df_mergedPlatLoc4.groupby(["LAT", "LON", "ALT"], as_index=False)["label_numeric"].value_counts(normalize=True)
df_mergedPlatLoc4_count = df_mergedPlatLoc4_count.rename(columns={"label_numeric": "emotion_value"})
conditions = [
    df_mergedPlatLoc4_count["emotion_value"] == "0",
    df_mergedPlatLoc4_count["emotion_value"] == "1",
    df_mergedPlatLoc4_count["emotion_value"] == "2",
    df_mergedPlatLoc4_count["emotion_value"] == "3",
    df_mergedPlatLoc4_count["emotion_value"] == "4",
    df_mergedPlatLoc4_count["emotion_value"] == "5",
    df_mergedPlatLoc4_count["emotion_value"] == "6"
]
values = ["anger", "disgust", "fear", "joy", "neutral", "sadness", "surprise"]
df_mergedPlatLoc4_count["emotion_label"] = np.select(conditions, values, default="Unknown")
df_mergedPlatLoc4_confScore = df_mergedPlatLoc4.groupby(["LAT", "LON", "ALT"], as_index=False)["score"].mean()
df_mergedPlatLoc4_confScore = df_mergedPlatLoc4_confScore.rename(columns={"score": "average_confidence_score"})
df_mergedPlatLoc4_count_conf = pd.merge(df_mergedPlatLoc4_count, df_mergedPlatLoc4_confScore, on=("LAT", "LON", "ALT"), how="left")
print(df_mergedPlatLoc4_count_conf)
df_mergedPlatLoc4_count_conf.to_csv("Results_EmotionsPlateauLocations_Proportion.csv", encoding="utf-8-sig", index=False, header=True, mode="w+")
```
